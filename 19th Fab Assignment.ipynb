Q3. What is App routing in Flask? Why do we use app routes?

->> App routing in Flask refers to the process of mapping URLs to their corresponding view functions or handlers. It allows Flask applications       to respond to specific URLs with the appropriate logic and data.

* Reasons of using app routers

1.Modular organization: App routing allows us to break up our application into smaller, more manageable modules. We can define different routes   for different parts of our application and group related functionality together.

2.Dynamic routing: Flask's app routers support dynamic routing, which allows us to handle variable parts of a URL. For example, we can handle     URLs that include user IDs or product names as variables.

3.Clean URLs: App routing allows us to create clean and meaningful URLs that are easy to read and remember. This makes our web application more   user-friendly and easier to navigate.

4.Separation of concerns: App routing helps us to separate the concerns of handling incoming requests and generating responses from the rest of   our application logic. This makes our code more modular and easier to maintain.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.



->> In Flask url_for() function is used for URL Building.

```python
from flask import Flask
from flask import url_for

app=Flask(__name__)

@app.route('/')
def index():
    return 'Hello World!'

@app.route('/user/<username>')
def show_user_profile(username):
    return f'User {username}'

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return f'Post {post_id}'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('show_user_profile', username='john'))
    print(url_for('show_post', post_id=5))

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
